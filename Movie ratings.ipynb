```python
movies = pd.read_csv("Movie-Ratings.csv")
```

```python
movies.columns = ['Film','Genre','CriticRating','AudienceRating','Budget','Year']
```

```python
movies.head()
```

```python
movies.Year = movies.Year.astype('category')
movies.Genre = movies.Genre.astype('category')
movies.Film = movies.Film.astype('category')
```

```python
sns.set_style("dark",{"axes.facecolor":"black"})
f, axes = plt.subplots(2,2, figsize=(10,10))

#Plot 1
k1= sns.kdeplot(movies.Budget, movies.CriticRating, ax = axes[0,0], shade=True, shade_lowest=True, cmap='inferno')
k1b = sns.kdeplot(movies.Budget, movies.CriticRating, ax = axes[0,0], cmap='cool')

#Plot 2
k2 = sns.kdeplot(movies.Budget, movies.AudienceRating, ax = axes[0,1],shade=True, shade_lowest=True, cmap='inferno')
k2b = sns.kdeplot(movies.Budget, movies.AudienceRating, ax = axes[0,1], cmap='cool')

#Plot 3
v = sns.violinplot(data = movies, x='Genre', y = 'CriticRating',ax=axes[1,0], palette='YlOrRd')

#Plot 4
m1 = sns.kdeplot(movies.CriticRating, movies.AudienceRating, shade=True, shade_lowest=False, cmap='cool_r', ax=axes[1,1])
m2 = sns.kdeplot(movies.CriticRating, movies.AudienceRating,  ax=axes[1,1],cmap='gray_r')
k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
list1 = list()
mylabel=list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].Budget)
    mylabel.append(gen) 
    
sns.set_style("whitegrid")
fig, axes = plt.subplots()
fig.set_size_inches(11.7,8.7)
plt.title("Movie Budget Distribution", fontsize=35)
plt.hist(x=list1,stacked=True, label=mylabel, bins=30)
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
plt.xlabel(s="Budget", fontsize=25, color="green")
plt.ylabel(s="Movies", fontsize=25, color="red")
plt.legend(fontsize=11, frameon=True, fancybox=True)
plt.show()
```
